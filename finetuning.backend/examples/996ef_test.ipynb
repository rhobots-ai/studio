```python
# Step 1: Load pickle
pickle_file_path = '996ef_test3_data_ready_inv_20_df (2).pkl'
df = pd.read_pickle(pickle_file_path)

# Step 2: Create filtered DataFrame
df_filtered = df[['final_input', 'final_ground_truth']].copy()
instructions = """ 
                       You are an accountant looking at invoices from various sellers. You need to extract the following values. It is critical that you find the EXACT text as it appears in the invoice and do not modify, convert, or generate any values.
                       If a field is not found in the invoice, return null for that field.
                       
                       Return a JSON object with these exact field names:
                       - invoice_no (string if found else null if not found)
                       - invoice_date (DD-MM-YYYY format if found else null if not found)
                       - amount (integer only, no commas if found else null if not found)
                       - buyer_gstin (15 chars: 13 alphanumeric + Z + 1 alphanumeric, regex pattern [0-9A-Z]{13}Z[0-9A-Z] if found else null if not found)
                       - seller_gstin (15 chars: 13 alphanumeric + Z + 1 alphanumeric, regex pattern [0-9A-Z]{13}Z[0-9A-Z] if found else null if not found)
                       
                       Do not convert dates, add formatting to amounts, or modify any text. Return exactly what you see.
            
"""
df_filtered['final_instruction'] = instructions

# Step 3: Clean up whitespace in final_input
import re
def clean_input(text):
    return re.sub(r'\s+', ' ', text).strip() if isinstance(text, str) else text

df_filtered['final_input'] = df_filtered['final_input'].apply(clean_input)
df_filtered['final_instruction'] = df_filtered['final_instruction'].apply(clean_input)

# Step 4: Reorder and export
df_filtered = df_filtered[['final_instruction', 'final_input', 'final_ground_truth']]
df_filtered = df_filtered.sample(n=7000, random_state=42)


records = df_filtered.to_dict(orient='records')

with open('cleaned_data.json', 'w', encoding='utf-8') as f:
    json.dump(records, f, ensure_ascii=False, indent=4)
```
